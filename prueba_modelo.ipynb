# Prueba del modelo

#### Usaremos el modelo creado para hacer predicciones en un nuevo conjunto de datos.

```python
# Cargar librerias
import pandas as pd
from datetime import timedelta
#from geopy.distance import geodesic
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

#### Definición de los tipos de datos con base en el documento de descripción.

```python
dtypes = {
    "trip_id": "string", 
    "duration": "int8", 
    "start_time": "string", # A datetime.
    "end_time":"string", #  A datetime.
    "start_lat":"float32", 
    "start_lon":"float32", 
    "end_lat":"float32", 
    "end_lon":"float32", 
    "bike_id":"string", 
    "trip_route_category":"string", 
    "start_station":"string", 
    "end_station":"string", 
}
```

```python
# Cargar los datos
df = pd.read_csv('Data/test_set.csv',dtype=dtypes)
```

```python
df
```

#### Transformar datos para poder alimentarlos al modelo

Similares a los que hicimos en el proceso de EDA y de construcción del modelo

```python
df.isna().sum()
```

```python
df['start_time'] = pd.to_datetime(df['start_time'], format='mixed')
df['end_time'] = pd.to_datetime(df['end_time'], format='mixed')
```

#### Variable duration

```python
df[df['duration']<0]
```

```python
df['new_duration'] = df['end_time'] - df['start_time']
```

```python
df.loc[df['new_duration'] <= timedelta(minutes=0), 'new_duration'] = df['duration'].apply(lambda x: timedelta(minutes=x))
```

#### Imputar valores de duration

```python
df['duration']= df['new_duration'].dt.seconds / 60
```

```python
moda = df['duration'].mode()[0]
df.loc[df['duration'] <= 0, 'duration'] = moda
```

```python
df.isna().sum()
```

#### Generación de variables

```python
df['start_hour'] = df['start_time'].dt.hour  
df['start_day'] = df['start_time'].dt.day
df['start_month'] = df['start_time'].dt.month
df['start_year'] = df['start_time'].dt.year
df['start_weekday'] = df['start_time'].dt.weekday  
```

```python
# Transformar variables categóricas
cat_columns = ['start_station', 'end_station', 'trip_route_category']
for col in cat_columns:
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
features = ['start_station','end_station','start_hour','start_year', 'start_month', 'start_day','start_weekday', 'trip_route_category', 'duration']
X = df[features]
```

#### Datos para usar con el modelo

```python
X
```

#### Cargar el modelo

```python
import joblib

model = joblib.load('Modelo/lightgbm_model.pkl')

# Verificar
print(model)
```

#### Correr el modelo para generar las predicciones de la variable objetivo

```python
y_pred = model.predict(X)
```

```python
# Retransformar nombre de clases de variable objetivo
label_mapping = {'Annual Pass': 0, 'Flex Pass': 1, 'Monthly Pass': 2, 'One Day Pass': 3, 'Walk-up': 4}
inverse_mapping = {v: k for k, v in label_mapping.items()}
y_pred_labels = [inverse_mapping[pred] for pred in y_pred]
```

```python
# Asignar variable objetivo
df['passholder_type']=y_pred_labels
```

```python
# Ver resultados
df[['trip_id','passholder_type']]
```

```python
# Guardar datos como csv para evaluación en Kaggel
df[['trip_id','passholder_type']].to_csv('Data/set_final.csv', index=False)
```
